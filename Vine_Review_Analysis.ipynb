```python
# Import Dependancies
import pandas as pd
```

```python
# import and make dataframe from vine_table.csv 
vine_df = pd.read_csv("sql_tables/vine_table.csv")
vine_df.dropna()
vine_df.head()
```

```python
#Create new dataframe keeping only the rows with >=20 "total_votes"
vine_votes_df = vine_df[vine_df['total_votes']>=20]
vine_votes_df
```

```python
#Create a new Dataframe from vine_votes_df to keep rows where ("helpful_votes/total_votes") >= .5
vine_helpful_df = vine_votes_df[vine_votes_df['helpful_votes']/vine_votes_df['total_votes'] >= .5]
vine_helpful_df.head()
```

```python
# Create Dataframe where that retrievs all rows with "vine" == "Y"
vine_paid_df = vine_helpful_df[vine_helpful_df['vine'] == 'Y']
vine_paid_df.head()
vine_paid_df.count()
```

```python
# Create Dataframe where that retrievs all rows with "vine" == "N"
vine_unpaid_df = vine_helpful_df[vine_helpful_df['vine'] == 'N']
vine_unpaid_df.head()
vine_unpaid_df.count()
```

```python
# Calculate the # of Reviews for Vine Program members, the # of 5 star reviews and the % 5 star reviews

vine_paid_cnt  = vine_paid_df['review_id'].count()
vine_paid_5star = vine_paid_df['review_id'][vine_paid_df['star_rating'] == 5].count()
vine_paid_5star_percent = vine_paid_5star/vine_paid_cnt * 100

print(f"""There were {vine_paid_cnt} reviews from Vine Members, with {vine_paid_5star} 5-star reviews or 
                  {vine_paid_5star_percent:.1f}% of all Vine Member reviews.""")
```

```python
# Calculate the # of Reviews for non-Vine Program members, the # of 5 star reviews and the % 5 star reviews

vine_unpaid_cnt  = vine_unpaid_df['review_id'].count()
vine_unpaid_5star = vine_unpaid_df['review_id'][vine_unpaid_df['star_rating'] == 5].count()
vine_unpaid_5star_percent = vine_unpaid_5star/vine_unpaid_cnt * 100

print(f"""There were {vine_unpaid_cnt} reviews from non-Vine Members, with {vine_unpaid_5star} 5-star reviews
            or {vine_unpaid_5star_percent:.1f}% of all non-Vine Member reviews.""")
```
